Home Price Prediction Using Simple Linear Regressions

```python
import pandas as pd
df = pd.read_csv("./dataset/homeprices.csv")
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.xlabel("area")
plt.ylabel("price")
plt.title("Housing Price")
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
new_df = df.drop('price',axis='columns')
new_df
```

```python
from sklearn import linear_model

#Create Linear Regression Object
reg = linear_model.LinearRegression()
reg.fit(new_df,df.price)
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
reg.predict([[3300]])
```

```python
plt.scatter(df.area,df.price, color='red',marker='+')
plt.plot(new_df, reg.predict(new_df),color='blue')
plt.title("Housing Price Prediction") # stating the title of the graph
plt.xlabel("Area") # adding the name of x-axis
plt.ylabel("Price") # adding the name of y-axis
plt.show()
```
